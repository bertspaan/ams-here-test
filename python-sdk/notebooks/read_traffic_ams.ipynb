```python
import geohash
from ipyleaflet import Heatmap
import pprint



import nagini
from nagini.core.hls.map import HLSMap
from nagini.utils.geo import Point, BoundingBox
from nagini.utils.tiling import tile_ids_from_bounding_box

olp = nagini.resource('olp')
cat = olp.catalog_by_hrn('hrn:here:data::olp-amsterdam:nld-traffic-decoded')
layer = cat.layer_by_id('traffic')
boundingbox = BoundingBox(Point(latitude=52.33673, longitude=5.08022), Point(latitude=52.5, longitude=4.65416) )
zoom_level = 12
partition_ids = tile_ids_from_bounding_box(boundingbox, zoom_level)
tile_ids = list(partition_ids)
#print (tile_ids)
blob = next(layer.read_partitions(partition_ids=tile_ids,version = 0))
csvfile = blob.decode('utf8')
#'23605384'

print (csvfile)



```
